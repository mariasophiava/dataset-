```python
import pandas as pd
import numpy as np
```

```python
data = {'company': ['Ford', 'Ford','VMW','AUDI', 'AUDI','Ford'],
      'Price': [20,22,30,34,25,28],
      'Weight': [55,60,70,28,30,80],
      'Length': [30,30,40,50,20, pd.NaT],
      'Made': ['China','Mexico','China','Mexico',np.nan,np.nan]
      }
df = pd.DataFrame(data = data)
df
```

```python
ndf=df.copy()
ndf
```
