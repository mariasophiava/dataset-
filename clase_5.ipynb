```python
favs=['Age', 'Gender', 'Price']
ndf = df[favs]
ndf.dropna(how = 'all',
           axis = 1,
           inplace = True)
ndf
```

##Which age group is satisfied with price factor?

```python
ndf.head(2)
```

```python
def cust_cls(x):
  if (0<=x<=20):
    return '<=20'
  elif (20<x<=30):
    return '21-30'
  elif (30<x<=40):
    return '31-40'
  else:
    return '>40'
```

```python
cust_cls(-5)
```

```python
NEWF= lambda tbl:cust_cls (tbl['Age'])
df['AgeGP']= ndf.apply (NEWF, axis =1 )
ndf
```

```python
ndf
```

```python
ndf ['AgeGP']= ndf.apply (NEWF, axis = 1)
```

```python
MY_PIVT = pd.pivot_table(ndf,
               index = 'AgeGP',
               aggfunc = 'count',
               values ='Price',
               )
MY_PIVT
```

```python
MY_PIVT.rename({'Price':'Numero'},
               axis = 1,
               inplace = True)
MY_PIVT.reset_index(inplace = True)
MY_PIVT
```

```python
import matplotlib.pyplot as plt
```

```python
explode_size = [0.03,0.03,0.04,0.03]
color_favs=['gold','yellowgreen','coral','skyblue','pink']
MY_PIVT['Numero'].plot(kind='pie',
                       figsize = [15, 6],
                       startangle =90,
                       autopct ='%1.1f%%',
                       shadow =True,
                       labels =None,
                       pctdistance =1.1,
                       explode = explode_size,
                       colors=color_favs
                       );
plt.legend(labels = MY_PIVT.AgeGP, bbox_to_anchor = [1.0,1.0])
```
